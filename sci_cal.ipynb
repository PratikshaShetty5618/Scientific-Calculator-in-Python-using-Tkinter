```python
from tkinter import *
import math


class Calc():
    def __init__(self):
    #to store the result
        self.total = 0
        self.ans=""
    #to store the current input by the user
        self.current = ""
    #to check for requirement of n
        self.new_num = True
    #to check whether any operator is not served
        self.op_pending = False
    #to store the operator given
        self.op = ""
    #to check for pending total operation 
        self.eq = False

    def num_press(self, num):
        self.eq = False
        temp = text_box.get()
        temp2 = str(num)
        if self.new_num:
            self.current = temp2
            self.new_num = False
        else:
            if temp2 == '.':
                if temp2 in temp:
                    return
            self.current = temp + temp2
        self.display(self.current)


    def calc_total(self):
        self.eq = True
        self.current = float (float(self.current))
        if self.op_pending == True:
            self.do_sum()
        else:
            self.total = float(text_box.get())

    def display(self, value):
        text_box.delete(0, END)
        text_box.insert(0, value)

    def do_sum(self):
        if self.op == "add":
            self.total += self.current
        if self.op == "minus":
            self.total -= self.current
        if self.op == "times":
            self.total *= self.current
        if self.op == "divide":
            self.total /= self.current
        if self.op == "raise":
            self.total = self.total ** self.current
        if self.op == "rootof":
            self.total = self.total ** (1/self.current)
        if self.op == "fact":
            self.total=int(text_box.get())
            self.total=math.factorial(self.total)
        if self.op == "ln":
            self.total = math.log(self.total)
        if self.op == "sqrt":
            self.total=math.sqrt(self.total)
        if self.op == "sine":
            self.total=math.sin(self.total)
        if self.op == "cosine":
            self.total = math.cos(self.total)
        if self.op == "tangent":
            self.total = math.tan(self.total)
            if(self.total>=int(16331239353195370.0)):
                self.total="Math Error"
        if self.op == "exp":
            self.total = math.exp(self.total)
        if self.op == "inv":
            self.total = 1/self.total
        if self.op == "invsine":
            self.total = math.asin(self.total)
        if self.op == "invcosine":
            self.total = math.acos(self.total)
        if self.op == "invtan":
            self.total = math.atan(self.total)
        if self.op == "sinh":
            self.total = math.sinh(self.total)
        if self.op == "cosh":
            self.total = math.cosh(self.total)
        if self.op == "tanh":
            self.total = math.tanh(self.total)
        if self.op == "e":
            self.total = math.e
        if self.op == "pi":
            self.total = math.pi
        if self.op == "rtd":
            self.total = math.degrees(self.total)
        if self.op == "dtr":
            self.total = math.radians(self.total)
               
        self.flag=1
        self.new_num = True
        self.op_pending = False
        self.ans=str(self.total)
        
        if(self.ans[-1]=='0'):
            self.display(int(self.total))
            self.flag=0
        if(self.flag==1):
            self.display(self.total)
        if(self.ans[-4]=='e'):
            self.total=0
            self.display(self.total)
        if(float(self.ans)==0.9999999999999999):
            self.total=1
            self.display(self.total)
            
    def operation(self, op):
        self.current = float(self.current)
        if self.op_pending:
            self.do_sum()
        elif not self.eq:
            self.total = self.current
        self.new_num = True
        self.op_pending = True
        self.op = op
        self.eq = False

    def clear(self):
        self.eq = False
        self.current = "0"
        self.display(0)
        self.new_num = True

    def all_clear(self):
        self.clear()
        self.total = 0

    def sign(self):
        self.eq = False
        self.current = -(float(text_box.get()))
        self.ans=str(self.current)
        if (self.ans[-1]=='0'):
            self.display(int(self.current))
        else:
            self.display(self.current)

#to create object of class
sum1 = Calc()
#to create Tk root widget(fot title and other decoration of window)
root = Tk()

calc = Frame(root)
calc.grid()

root.title("Calculator")


#insert text box for input and output
text_box = Entry(calc, justify=RIGHT,width=30,font="Arial 16 bold")
text_box.grid(row = 0, column = 0,columnspan = 8,padx=30, pady = 30)
text_box.insert(0, "0")

#to enter numbers from 1 to 9
numbers = "789456123"
i = 0
bttn = []
#j is used for arranging in row(i.e start from 1st row upto 3rd row)
for j in range(1,4):
    #k is used for arranging in column(i.e start from 0th col to 2nd col)
    for k in range(3):
        
        #append is used to insert buttons from recent end position
        bttn.append(Button(calc,height =2,width=4,padx=10, pady = 10, text = numbers[i],font="Arial 18 bold"))
        
        bttn[i]["bg"]= "yellow"
        bttn[i].grid(row = j, column = k,padx=1,pady=1)
        bttn[i]["command"] = lambda x = numbers[i]: sum1.num_press(x)
        i += 1

bttn_0 = Button(calc,height =2,width=4,padx=10, pady = 10, text = "0",bg="yellow",font="Arial 16 bold")
bttn_0["command"] = lambda: sum1.num_press(0)
bttn_0.grid(row = 4, column = 0,  padx=1, pady = 1)

div = Button(calc,height =2,width=4,padx=10, pady = 10, text = "/",bg="light blue",font="Arial 16 bold")
div["command"] = lambda: sum1.operation("divide")
div.grid(row = 1, column = 3, padx=1, pady = 1)

mult = Button(calc,height =2,width=4,padx=10, pady = 10, text = "*",bg="light blue",font="Arial 16 bold")
mult["command"] = lambda: sum1.operation("times")
mult.grid(row = 2, column = 3,  padx=1, pady = 1)

minus = Button(calc,height =2,width=4,padx=10, pady = 10, text = "-",bg="light blue",font="Arial 16 bold")
minus["command"] = lambda: sum1.operation("minus")
minus.grid(row = 3, column = 3, padx=1, pady = 1)

add = Button(calc,height =2,width=4,padx=10, pady = 10, text = "+",bg="light blue",font="Arial 16 bold")
add["command"] = lambda: sum1.operation("add")
add.grid(row = 4, column = 3,  padx=1, pady = 1)

power = Button(calc, height=2,width=4,padx=10,pady=10,text="x^y",bg="green",font="Arial 16 bold")
power["command"] = lambda: sum1.operation("raise")
power.grid(row=2,column = 4,padx=1,pady=1)

rootof = Button(calc, height=2, width=4, padx=10, pady=10, text="x root of y", bg = "green",font="Arial 16 bold")
rootof["command"] = lambda: sum1.operation("rootof")
rootof.grid(row=2, column=5, padx=1, pady=1)

fact = Button(calc, height=2, width=4, padx=10, pady=10, text="!",bg="green",font="Arial 16 bold")
fact["command"] = lambda: sum1.operation("fact")
fact.grid(row=3,column=4, padx=1, pady=1)

loge = Button(calc, height=2, width=4, padx=10, pady=10, text="ln",bg="green",font="Arial 16 bold")
loge["command"] = lambda: sum1.operation("ln")
loge.grid(row=3, column=5, padx=1, pady=1)

sqrt = Button(calc, height=2, width=4, padx=10, pady=10, text="sqrt",bg="green",font="Arial 16 bold")
sqrt["command"]= lambda: sum1.operation("sqrt")
sqrt.grid(row=4, column=4, padx=1 , pady=1)

sine = Button(calc, height=2,width=4, padx=10,pady=10, text = "sin" , bg= "green",font="Arial 16 bold")
sine["command"]=lambda: sum1.operation("sine")
sine.grid(row=5,column=0,padx=1,pady=1)

cosine = Button(calc, height=2,width=4, padx=10,pady=10, text = "cos" , bg= "green",font="Arial 16 bold")
cosine["command"]=lambda: sum1.operation("cosine")
cosine.grid(row=5,column=1,padx=1,pady=1)

tangent = Button(calc, height=2,width=4, padx=10,pady=10, text = "tan" , bg= "green",font="Arial 16 bold")
tangent["command"]=lambda: sum1.operation("tangent")
tangent.grid(row=5,column=2,padx=1,pady=1)

invsine=Button(calc,height=2,width=4,padx=10,pady=4,text="sin^-1",bg="green",font="Arial 16 bold")
invsine["command"]=lambda: sum1.operation("invsine")
invsine.grid(row=6,column=0,padx=1,pady=1)

invcosine=Button(calc,height=2,width=4,padx=10,pady=4,text="cos^-1",bg="green",font="Arial 16 bold")
invcosine["command"]=lambda: sum1.operation("invcosine")
invcosine.grid(row=6,column=1,padx=1,pady=1)

invtan=Button(calc,height=2,width=4,padx=10,pady=4,text="tan^-1",bg="green",font="Arial 16 bold")
invtan["command"]=lambda: sum1.operation("invtan")
invtan.grid(row=6,column=2,padx=1,pady=1)

exponent = Button(calc, height=2, width=4, padx=10, pady=10, text='e^x', bg="green",font="Arial 16 bold")
exponent["command"]=lambda: sum1.operation("exp")
exponent.grid(row=5,column=3,padx=1,pady=1)

inv = Button(calc, height=2, width=4, padx=10, pady=10, text="1/x", bg="green",font="Arial 16 bold")
inv["command"] = lambda: sum1.operation("inv")
inv.grid(row=5,column=4,padx=1,pady=1)

point = Button(calc,height =2,width=4,padx=10, pady = 10, text = ".",bg="white",font="Arial 16 bold")
point["command"] = lambda: sum1.num_press(".")
point.grid(row = 4, column = 1, padx=1, pady = 1)

neg= Button(calc,height =2,width=4,padx=10, pady = 10, text = "+/-",bg="white",font="Arial 16 bold")
neg["command"] = sum1.sign
neg.grid(row = 4, column = 2,  padx=1, pady = 1)

sinh=Button(calc,height=2,width=4,padx=10,pady=4,text="sinh",bg="green",font="Arial 16 bold")
sinh["command"]=lambda: sum1.operation("sinh")
sinh.grid(row=6,column=3,padx=1,pady=1)

cosh=Button(calc,height=2,width=4,padx=10,pady=4,text="cosh",bg="green",font="Arial 16 bold")
cosh["command"]=lambda: sum1.operation("cosh")
cosh.grid(row=6,column=4,padx=1,pady=1)

tanh=Button(calc,height=2,width=4,padx=10,pady=4,text="tanh",bg="green",font="Arial 16 bold")
tanh["command"]=lambda: sum1.operation("tanh")
tanh.grid(row=6,column=5,padx=1,pady=1)

dtr=Button(calc,height=2,width=13,padx=10,pady=4,text="DEG TO RAD",bg="green",font="Arial 16 bold")
dtr["command"]=lambda: sum1.operation("dtr")
dtr.grid(row=7,column=4,columnspan=2,rowspan=1,padx=1,pady=1)

rtd=Button(calc,height=2,width=13,padx=10,pady=4,text="RAD TO DEG",bg="green",font="Arial 16 bold")
rtd["command"]=lambda: sum1.operation("rtd")
rtd.grid(row=7,column=2,columnspan=2,rowspan=1,padx=1,pady=1)

pi=Button(calc,height=2,width=4,padx=10,pady=4,text="pi",bg="green",font="Arial 16 bold")
pi["command"]=lambda: sum1.operation("pi")
pi.grid(row=7,column=1,padx=1,pady=1)

e=Button(calc,height=2,width=4,padx=10,pady=4,text="e",bg="green",font="Arial 16 bold")
e["command"]=lambda: sum1.operation("e")
e.grid(row=7,column=0,padx=1,pady=1)

clear = Button(calc,height =2,width=4,padx=10, pady = 10, text = "C",bg="white",font="Arial 16 bold")
clear["command"] = sum1.clear
clear.grid(row = 1, column = 4,  padx=1, pady = 1)

all_clear = Button(calc,height =2,width=4,padx=10, pady = 10, text = "AC",bg="white",font="Arial 16 bold")
all_clear["command"] = sum1.all_clear
all_clear.grid(row = 1, column = 5, padx=1, pady = 1)

equals = Button(calc,height =6,width=4,padx=10, pady = 10, text = "=",bg="green",font="Arial 16 bold")
equals["command"] = sum1.calc_total
equals.grid(row = 4, column = 5,columnspan=1,rowspan=2,padx=1, pady = 1)



root.mainloop()
```
